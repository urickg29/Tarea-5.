```python
class Array:
  def __init__( self , tam ):
    self.__tamanio = tam
    self.__data = [ 0 for x in range(self.__tamanio) ]

  def to_string( self ):
    info = ""
    for dato in self.__data:
      info = info + str(dato) +"|"
    return info

  def get_item( self , index ):
    return self.__data[index]

  def get_length( self ):
    return len( self.__data )

  def set_item( self , dato , index):
    # hace falta revisar limites
    if index >= 0 and index < self.get_length():
      self.__data[index] = dato
    else:
      print("Error: indice fuera de rango")

  def clear( self , dato ):
    self.__data = [ dato for x in range(self.__tamanio) ]
```

```python
class Trabajador:
    def __init__(self, nt, nom, pat, mat, he, sb, ai) :
        self.__numero_trabajador=nt
        self.__nombre=nom
        self.__apellido_paterno=pat
        self.__apellido_materno=mat
        self.__horas_extras=he
        self.__sueldo_base=sb
        self.__anio=ai
    
    def get_numero_trabajador (self) :
        return self.__numero_trabajador

    def set_numero_trabajador(self, nt) :
        self.__numero_trabajador=nt 

    def get_nombre (self) :
        return self.__nombre

    def set_nombre(self, nom) :
        self.__nombre=nom

    def get_apellido_paterno(self) :
        return self.__apellido_paterno

    def set_apellido_paterno(self, pat) :
        self.__apellido_paterno = pat
    
    def get_apellido_materno(self) :
        return self.__apellido_materno

    def set_apellido_materno(self, mat) :
        self.__apellido_materno = mat
    
    def get_horas_extras(self) :
        return self.__horas_extras

    def set_horas_extras(self, he) :
        self.__horas_extras = he
    
    def get_sueldo_base(self) :
        return self.__sueldo_base

    def set_sueldo_base(self, sb) :
        self.__sueldo_base = sb

    def get_anio(self) :
        return self.__anio

    def set_anio(self, ai) :
        self.__anio = ai
    
    def to_String (self) : 
        return "Número de trabajador : " + str(self.__numero_trabajador) + "\nNombre : " + self.__nombre + "\nApellido Paterno : " + self.__apellido_paterno + "\nApellido Materno : " + self.__apellido_materno + "\nHoras Extras : " + str(self.__horas_extras) + "\nSueldo Base : " + str(self.__sueldo_base) + "\nAnio de Ingreso : " + str(self.__anio) + "\nSueldo ya calculado : " + str(self.calcular_sueldo_mensual())
    
    def calcular_sueldo_mensual (self) :
        anios = 2021 - self.get_anio()
        sueldo = self.get_sueldo_base()
        #agregar horas extras
        sueldo+= self.get_horas_extras() * 276.5
        #agregar prestacion por antiguedad
        sueldo=sueldo + (sueldo*anios*0.03)
        return sueldo

    def calcular_antiguedad(self) :
        return 2021 - self.get_anio()
```

```python
class NominaADT:
    def __init__(self, ruta_archivo) :
        archivo = open(ruta_archivo, 'rt', encoding='latin-1')
        empleados = archivo.readlines()
        del empleados[0]
        for index in range(len(empleados)) :
            empleados[index] = empleados[index].strip().split(',')
        self.__data = Array(len(empleados))
        cont=0
        for emp in empleados :
            tmp = Trabajador(int(emp[0]), emp[1], emp[2], emp[3], int(emp[4]), int(emp[5]), int(emp[6]))
            self.__data.set_item(tmp, cont)
            cont+=1
    
    def buscar_menor_antiguo(self) :
        print("Los trabajadores con menos antiguedad son:\n")
        anios=[]
        for index in range(self.__data.get_length()) :
            anios.append(self.__data.get_item(index).get_anio())
        anios.sort()
        anios.reverse()
        menor_antiguo = anios[0]
        for index in range(self.__data.get_length()) :
            if menor_antiguo == self.__data.get_item(index).get_anio() :
                print(self.__data.get_item(index).to_String() +"\n")

    
    def buscar_mas_antiguo(self) :
        print("Los trabajadores con mayor antiguedad son:\n")
        anios=[]
        for index in range(self.__data.get_length()) :
            anios.append(self.__data.get_item(index).get_anio())
        anios.sort()
        mayor_antiguo = anios[0]
        for index in range(self.__data.get_length()) :
            if mayor_antiguo == self.__data.get_item(index).get_anio() :
                print(self.__data.get_item(index).to_String() +"\n")

    def calcular_sueldos(self) :
        print("El calculo de los sueldos es:\n")
        for index in range(self.__data.get_length()) :
            print(self.__data.get_item(index).to_String() + "\n")
```
